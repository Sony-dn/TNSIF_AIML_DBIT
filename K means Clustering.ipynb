```python
import os
os.environ["OMP_NUM_THREADS"] = "1"

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv(r"C:\Users\dnson\Downloads\student_kmeans_dataset.csv")
X = df[["Study_Hours", "Exam_Score","Student_ID"]]
kmeans3 = KMeans(n_clusters=3, random_state=42)
df["Cluster_K3"] = kmeans3.fit_predict(X)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df["Study_Hours"], df["Exam_Score"], 
            c=df["Cluster_K3"], cmap="viridis", s=100, edgecolor="k")

centers = kmeans3.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c="red", s=200, marker="X", label="Centroids")

plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("K-Means Clustering (K=3)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.show()
```

```python
kmeans4 = KMeans(n_clusters=4, random_state=42)
df["Cluster_K4"] = kmeans4.fit_predict(X)

plt.figure(figsize=(8,6))
plt.scatter(df["Study_Hours"], df["Exam_Score"], 
            c=df["Cluster_K4"], cmap="viridis", s=120, edgecolors="black")

centers = kmeans4.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], 
            c="red", s=200, alpha=0.7, marker="X", label="Centroids")

plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("K-Means Clusters (K=4)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.show()
```

```python
kmeans2 = KMeans(n_clusters=2, random_state=42)
df["Cluster_K2"] = kmeans2.fit_predict(X)

centroids = kmeans2.cluster_centers_

plt.figure(figsize=(8,6))
plt.scatter(df["Study_Hours"], df["Exam_Score"], 
            c=df["Cluster_K2"], cmap="viridis", s=100, label="Students")
plt.scatter(centroids[:,0], centroids[:,1], 
            c="red", marker="X", s=200, label="Centroids")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("K-Means Clustering (K=2)")
plt.grid(True)
plt.legend()
plt.show()
```
